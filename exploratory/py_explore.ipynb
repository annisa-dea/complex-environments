```python
import pandas as pd
import matplotlib.pyplot as plt

umap_df = pd.read_csv("results/evo/umap/umap_coords.csv", index_col=0)

colors = umap_df["30c"].map({0: "#cccccc", 1: "#1f77b4"})

fig, ax = plt.subplots(figsize=(5, 5))
ax.scatter(umap_df["UMAP_1"], umap_df["UMAP_2"],
           c=colors, s=1, alpha=0.5, rasterized=True)

ax.set_xlabel("UMAP 1")
ax.set_ylabel("UMAP 2")
ax.set_title("Osmolytes")
ax.spines[["top", "right"]].set_visible(False)

from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#cccccc', markersize=7, label='osmolytes = 0'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#1f77b4', markersize=7, label='osmolytes = 1'),
]
ax.legend(handles=legend_elements, frameon=False, fontsize=9)
plt.tight_layout()
plt.show()
```
