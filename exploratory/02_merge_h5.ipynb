Notebook is used to load .h5 data that comes out of the sequencer and merges them to one AnnData input file by the end

```python
import numpy as np
import pandas as pd
import scanpy as sc
import os
import anndata as ad
from pathlib import Path

sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_header()
sc.settings.set_figure_params(dpi=80, facecolor='white')

wd = Path('/Users/annisa/Box Sync/G7/all_nova_filtered/anc_h5')
conditions_list_os = sorted([f[:-3] for f in os.listdir(wd) if f.endswith(".h5")])
conditions_dict = dict.fromkeys(conditions_list_os)
```

```python

metadata = pd.read_csv("/Users/annisa/Box Sync/G7/scales_run_ipynb/ConditionsMatrixCombined.csv", index_col=0)
metadata = metadata[metadata["ancestry"] == 0] # select evolved only
```

```python
metadata
```

```python
for index, row in metadata.iterrows():
    adata = sc.read_10x_h5(wd / f"{index}.h5")
    adata.obs_names = [f"{index}_{bc}" for bc in adata.obs_names]
    adata.var_names_make_unique()
    for k, v in row.items():
        adata.obs[k] = v
    conditions_dict[index] = adata
```

```python
newadatas = ad.concat(conditions_dict, label="condition")
newadatas.obs_names.duplicated().sum() #check for 0
```

```python
#newadatas.var_names_make_unique()
newadatas.write("mergedh5_nova_anc_CRfiltered.h5ad")
```
